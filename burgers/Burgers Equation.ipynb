# Solving the Burgers Equation

## Import Relevant Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import sciann as sn
from sciann.utils.math import diff, sign, sin
```

## Define Inputs and Build the DNN

```python
# Variable defines the inputs 
x = sn.Variable('x')
t = sn.Variable('t')

# Functional constructs the neural network approximation
u = sn.Functional('u', [t,x], 8*[20], 'tanh')
```

## Set Up Optimization Problem with Boundary Conditions

```python
L1 = diff(u, t) + u*diff(u,x) - (0.01/np.pi)*diff(u, x, order=2)

TOL = 0.001
L2 = (1-sign(t - TOL)) * (u + sin(np.pi*x))
L3 = (1-sign(x - (-1+TOL))) * (u)
L4 = (1+sign(x - ( 1-TOL))) * (u)
```

## Build the PINN

```python
m = sn.SciModel([x, t], [L1, L2, L3, L4])
```

## Get data for x and t

```python
x_data, t_data = np.meshgrid(
    np.linspace(-1, 1, 100), 
    np.linspace(0, 1, 100)
)
```

## Train Model

```python
h = m.train(
    [x_data, t_data], 
    4*['zero'], 
    learning_rate=0.002, 
    epochs=5000,
    verbose = 0)
```

## Get Test Data and Predictions

```python
x_test, t_test = np.meshgrid(
    np.linspace(-1, 1, 200), 
    np.linspace(0, 1, 200)
)
u_pred = u.eval(m, [x_test, t_test])
```

## Plot Predictions

```python
fig = plt.figure(figsize=(3, 4))
plt.pcolor(x_test, t_test, u_pred, cmap='seismic')
plt.xlabel('x')
plt.ylabel('t')
plt.colorbar()
```

## Build New Model (using tanh for boundary conditions instead of sign)

```python
from sciann.utils.math import tanh
L5 = (1-tanh(t - TOL)) * (u + sin(np.pi*x))
L6 = (1-tanh(x - (-1+TOL))) * (u)
L7 = (1+tanh(x - ( 1-TOL))) * (u)
```

```python
m2 = sn.SciModel([x, t], [L1, L5, L6, L7])
```

```python
h2 = m2.train(
    [x_data, t_data], 
    4*['zero'], 
    learning_rate=0.002, 
    epochs=5000,
    verbose = 0)
```

```python
u_pred2 = u.eval(m2, [x_test, t_test])
```

```python
fig = plt.figure(figsize=(3, 4))
plt.pcolor(x_test, t_test, u_pred2, cmap='seismic')
plt.xlabel('x')
plt.ylabel('t')
plt.colorbar()
```
